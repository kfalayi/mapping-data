```python
import pandas as pd
import numpy as np
!pip install geopandas
import geopandas as gpd
import altair as alt
```

```python
pd.set_option('display.max_columns', None)
wildfire = pd.read_excel('https://brown-institute-assets.s3.amazonaws.com/Objects/pointsunknown/WildfireRiskToCommunities_County_Summary_2020-03-31.xlsx')
wildfire.head()
```

```python
#Group 'Expected Annual HU Exposed' by state
EA_unitExposed = wildfire.groupby(['STATEFP', 'STATE'])['Expected Annual HU Exposed'].sum().to_frame()
EA_unitExposed['Expected Annual HU Exposed'] = EA_unitExposed['Expected Annual HU Exposed'].round()
EA_unitExposed.head()
```

```python
#Group 'Total number of exposed HUs' by state
totalHU = wildfire.groupby(['STATEFP', 'STATE'])['Total number of exposed HUs'].sum().to_frame()
totalHU.head()
```

```python
#Merge both 'Total number of exposed HUs' and 'Expected Annual HU Exposed'  together
statesData = EA_unitExposed.merge(totalHU, on='STATEFP').reset_index()
statesData.head()
```

```python
#Create a dataframe of number of rows for each state
state_length = wildfire.STATEFP.value_counts()
state_length.head()
```

```python
#Group 'Fraction EAHUexposed-Direct' by state.
wildfire.groupby('STATEFP')['Fraction EAHUexposed-Direct'].sum()
```

```python
#convert 'Fraction EAHUexposed-Direct' by state back to percentage
stateFraction = wildfire.groupby('STATEFP')['Fraction EAHUexposed-Direct'].sum() / state_length * 100
stateFraction.head()
```

```python
stateFraction = stateFraction.to_frame().reset_index()
stateFraction.head()
```

```python
stateFraction = stateFraction.rename(columns={'index': 'STATEFP'})
```

```python
stateFraction = stateFraction.rename(columns={0: 'Fraction EAHUexposed-Direct'})
stateFraction.head(2)
```

```python
#convert the percentages to fractions
stateFraction['Fraction EAHUexposed-Direct'] = stateFraction['Fraction EAHUexposed-Direct'] / 100
```

```python
stateFraction.head()
```

```python
#merge with statesData
statesData = stateFraction.merge(statesData, on='STATEFP')
```

```python
statesData.head(2)
```

## Natural Breaks (Jenks)

```python
!pip install jenkspy
import jenkspy
```

```python
breaks = jenkspy.jenks_breaks(statesData['Fraction EAHUexposed-Direct'], nb_class=5)
breaks
```

```python
naturalBreaksLabels = []
for x in range(0,len(breaks)-1):
  naturalBreaksLabels.append(str(round(breaks[x]*100))+'% - ' + str(round(breaks[x+1]*100)) + '%')
print(naturalBreaksLabels)
```

```python
statesData['naturalBreaks'] = pd.cut(statesData['Fraction EAHUexposed-Direct'], breaks, labels=naturalBreaksLabels)
```

```python
statesMap = alt.Chart(statesData).mark_geoshape().encode(
    color=alt.Color('naturalBreaks:O',
      legend=alt.Legend(title='% of Expected Annual HUs directly Exposed'),
      scale=alt.Scale(scheme='lightorange',domain=naturalBreaksLabels)),
    tooltip=[alt.Tooltip('NAME',title='State'),
      alt.Tooltip('Fraction EAHUexposed-Direct',title='% of Expected Annual HUs directly Exposed',format='.0%')]
).project(
    type='albersUsa'
).properties(
    title='Percentage of Housing Units with Expected Annual Direct Exposure to Wildfire',
    width=700,
    height=450
).configure_view(
    strokeWidth=0
)

statesMap
```
